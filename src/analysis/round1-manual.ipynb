```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
reserve_prices = np.sqrt(np.random.random(10_000_000)) * 100 + 900
px.histogram(reserve_prices[:100_000]).show()
```

```python
lows = []
highs = []
profits = []
ppf_takens = []
ppf_totals = []

take_bid = [bid > reserve_prices for bid in range(900, 1001)]
total_fishes = len(reserve_prices)

for low in range(900, 1001):
    take_low = take_bid[low - 900]
    low_fishes_taken = take_low.sum()
    low_profit = low_fishes_taken * (1000 - low)

    for high in range(low, 1001):
        high_fishes_taken = (take_low ^ take_bid[high - 900]).sum()

        profit = low_profit + high_fishes_taken * (1000 - high)
        ppf_taken = profit / (low_fishes_taken + high_fishes_taken) if (low_fishes_taken + high_fishes_taken) > 0 else 0
        ppf_total = profit / total_fishes

        lows.append(low)
        highs.append(high)
        profits.append(profit)
        ppf_takens.append(ppf_taken)
        ppf_totals.append(ppf_total)

df = pd.DataFrame({"low": lows, "high": highs, "profit": profits, "ppf_taken": ppf_takens, "ppf_total": ppf_totals})
df
```

```python
df.sort_values("profit", ascending=False)
```

```python
px.scatter_3d(df, x="low", y="high", z="profit", color="profit").update_layout(margin=dict(l=0, r=0, t=20, b=0)).show()
```

```python
px.scatter_3d(df, x="ppf_taken", y="ppf_total", z="profit", color="profit").update_layout(margin=dict(l=0, r=0, t=20, b=0)).show()
```
